```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_gradient(x):
    return sigmoid(x) * (1 - sigmoid(x))

def leaky_relu(x, alpha=0.3):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_gradient(x, alpha=0.3):
    return np.where(x > 0, 1, alpha)

def tanh(x):
    return np.tanh(x)

def tanh_gradient(x):
    return 1 - np.tanh(x)**2

x = np.linspace(-10, 10, 500)
y_sigmoid = sigmoid(x)
y_sigmoid_grad = sigmoid_gradient(x)
y_leaky_relu = leaky_relu(x)
y_leaky_relu_grad = leaky_relu_gradient(x)
y_tanh = tanh(x)
y_tanh_grad = tanh_gradient(x)

fig = plt.figure(figsize=(16, 4))

# Plot Sigmoid and its Gradient
ax1 = fig.add_subplot(131)
ax1.plot(x, y_sigmoid, color='blue', label=r'$\phi$')
ax1.plot(x, y_sigmoid_grad, color='blue', linestyle='dashed', label=r"$\phi$ '")
ax1.set_title("Sigmoid Activation Function")
ax1.set_xlim(-10, 10)
ax1.set_ylim(-1.25, 1.25)
ax1.set_xlabel("x", fontweight='bold').set_rotation(0)
ax1.set_ylabel("y", fontweight='bold').set_rotation(0)
ax1.legend()
ax1.spines['left'].set_position('zero')
ax1.spines['bottom'].set_position('zero')
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
ax1.xaxis.set_label_coords(0.75, 0.4)
ax1.yaxis.set_label_coords(0.35, 0.68)
ax1.text(x=1.9, y=-0.9, s=r"$\phi(x) = \frac{1}{1 + e^{-x}}$", fontsize=14)

# Plot Leaky ReLU and its Gradient
ax2 = fig.add_subplot(132)
ax2.plot(x, y_leaky_relu, color='blue', label=r'$\phi$')
ax2.plot(x, y_leaky_relu_grad, color='blue', linestyle='dashed', label=r"$\phi$ '")
ax2.set_title(r"Leaky ReLU Activation Function $\alpha=0.3$")
ax2.set_xlim(-10, 10)
ax2.set_ylim(-1.25, 1.25)
ax2.set_xlabel("x", fontweight='bold').set_rotation(0)
ax2.set_ylabel("y", fontweight='bold').set_rotation(0)
ax2.legend()
ax2.spines['left'].set_position('zero')
ax2.spines['bottom'].set_position('zero')
ax2.spines['right'].set_color('none')
ax2.spines['top'].set_color('none')
ax2.xaxis.set_ticks_position('bottom')
ax2.yaxis.set_ticks_position('left')
ax2.xaxis.set_label_coords(0.75, 0.4)
ax2.yaxis.set_label_coords(0.35, 0.68)
ax2.text(x=1.9, y=-0.9, s=r"$\phi(x) = max(\alpha x, x)$", fontsize=14)

# Plot Tanh and its Gradient
ax3 = fig.add_subplot(133)
ax3.plot(x, y_tanh, color='blue', label=r'$\phi$')
ax3.plot(x, y_tanh_grad, color='blue', linestyle='dashed', label=r"$\phi$ '")
ax3.set_title("Tanh Activation Function")
ax3.set_xlim(-10, 10)
ax3.set_ylim(-1.25, 1.25)
ax3.set_xlabel("x", fontweight='bold').set_rotation(0)
ax3.set_ylabel("y", fontweight='bold').set_rotation(0)
ax3.legend()
ax3.spines['left'].set_position('zero')
ax3.spines['bottom'].set_position('zero')
ax3.spines['right'].set_color('none')
ax3.spines['top'].set_color('none')
ax3.xaxis.set_ticks_position('bottom')
ax3.yaxis.set_ticks_position('left')
ax3.xaxis.set_label_coords(0.75, 0.4)
ax3.yaxis.set_label_coords(0.35, 0.68)
ax3.text(x=1.9, y=-0.9, s=r"$\phi(x) = tanh(x)$", fontsize=14)

plt.subplots_adjust(wspace=0.2)
plt.savefig("activation_functions.png", dpi=350)
plt.show()
```
